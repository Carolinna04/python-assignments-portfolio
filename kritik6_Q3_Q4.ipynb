```python
#QUESTION 3 CODE
import numpy as np
import matplotlib.pyplot as plt

def normal_density(mean, variance, x):
    sigma = np.sqrt(variance) #i.e standard deviation
    # Calculate the normal density function
    return (1 / (np.sqrt(2 * np.pi * variance))) * np.exp(-(x - mean)**2 / (2 * variance))

# Numerical integration using trapezoidal rule
def integration(mean, variance, a, b, n=1000):
    x_values = np.linspace(a, b, n)
    y_values = [normal_density(mean, variance, x) for x in x_values]
    return np.trapz(y_values, x_values)

# probability that a man has height between 162cm and 190cm
mean = 171
variance = 7.12
a = 162
b = 190

probability = integration(mean, variance, a, b)
print(f"The probability that a randomly chosen man has a height between {a} cm and {b} cm is: {probability}")

# Plotting the normal distribution
x_values = np.linspace(140, 200, 400)
y_values = [normal_density(mean, variance, x) for x in x_values]

plt.figure(figsize=(8, 6))
plt.plot(x_values, y_values, label=f'µ = {mean}, σ² = {variance}')
plt.title('Normal Distribution of Male Height')
plt.xlabel('Height (cm)')
plt.ylabel('Density')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Question 4(a):
def uniform_density(a, b, x):
    # Uniform density function
    if a <= x <= b:
        return 1 / (b - a)
    else:
        return 0

# Define parameters for uniform distribution
a = 5
b = 15
x_values = np.linspace(0, 20, 400)  # Generate x values

# Compute y values (probability density)
y_values = [uniform_density(a, b, x) for x in x_values]

# Plot
plt.figure(figsize=(8, 6))
plt.plot(x_values, y_values, label=f'Uniform Distribution on [{a}, {b}]')
plt.title('Uniform Distribution')
plt.xlabel('x')
plt.ylabel('Density')
plt.xlim(0, 20)
plt.ylim(0, 1.1)  # To show the constant value
plt.axvline(x=a, color='r', linestyle='--', label=f'a = {a}')
plt.axvline(x=b, color='r', linestyle='--', label=f'b = {b}')
plt.legend()
plt.grid(True)
plt.show()
 
```

```python
# Question 4(b): 

def exponential_density(lambda_, x):
    # Exponential density function
    return lambda_ * np.exp(-lambda_ * x) if x >= 0 else 0

# Define the rate for the exponential distribution
lambda_ = 1 / 50  # Rate for pandemics (1 per 50 years)
x_values = np.linspace(0, 200, 400)  # Generate x values (time between pandemics)

# Compute y values (probability density)
y_values = [exponential_density(lambda_, x) for x in x_values]

# Plot
plt.figure(figsize=(8, 6))
plt.plot(x_values, y_values, label=f'Exponential Distribution with λ = {lambda_}')
plt.title('Exponential Distribution')
plt.xlabel('Time (years)')
plt.ylabel('Density')
plt.xlim(0, 200)
plt.ylim(0, 0.05)
plt.legend()
plt.grid(True)
plt.show()

#The result aligns with what was expected, since the rate 𝜆=1/50 and the expected value is 50 years.
```

```python
# Question 4(c):
def normal_density(mu, sigma_squared, x):
    # Normal density function
    return (1 / np.sqrt(2 * np.pi * sigma_squared)) * np.exp(-(x - mu)**2 / (2 * sigma_squared))

# Define parameters for normal distribution
mu = 171  # Mean height in cm
sigma_squared = 7.1  # Variance in cm^2
x_values = np.linspace(140, 200, 400)  # Generate x values (heights)

# Compute y values (probability density)
y_values = [normal_density(mu, sigma_squared, x) for x in x_values]

# Plot
plt.figure(figsize=(8, 6))
plt.plot(x_values, y_values, label=f'Normal Distribution: µ = {mu}, σ² = {sigma_squared}')
plt.title('Normal Distribution of Male Height')
plt.xlabel('Height (cm)')
plt.ylabel('Density')
plt.legend()
plt.grid(True)
plt.show()
```
