```python
 import sympy as sp
x, y = sp.symbols('x y')
#Follow cells will show exercises
```

```python
# (a) Define the function f(x,y) = e^x sin(y)+y^3. Compute the first partial derivatives ∂f/ ∂x and ∂f/ ∂y.

# Define a function
f = sp.exp(x)*sp.sin(y) + y**3

# Differentiate f with respect to x
df_dx = sp.diff(f, x)

# Differentiate f with respect to y
df_dy = sp.diff(f, y)

#compute partial derivatives 
print("∂f/∂x:", df_dx)
print("∂f/∂y:", df_dy)
```

```python
# (b)  Given the function g(x,y) = x2y +xy2, find the gradient vector and compute its magnitude at the point (1,−1).

# Define the function
g = x**2*y + x*y**2

# Compute the gradient ( which are the partial derivatives with respect to x and y)
grad = [sp.diff(g, var) for var in (x, y)]

# gradient at the point (1, -1)
grad_at_point = [g_subs.subs({x: 1, y: -1}) for g_subs in grad]

# Compute the magnitude of the gradient vector
magnitude = sp.sqrt(sum(g**2 for g in grad_at_point))

# Display the results
print("Gradient vector:", grad_at_point)
print("Magnitude of the gradient:", magnitude)
```

```python
# (c) For the function h(x,y) = ln(x2+y2), find the second partial derivatives ∂2h ∂y2 , and ∂2h ∂x∂y. 
# Discuss the symmetry of mixed partial derivatives.

# Define the function
h = sp.ln(x**2 + y**2)

# Finf first partial derivatives first
dh_dx = sp.diff(h, x)
dh_dy = sp.diff(h, y)

# Now find second partial derivatives:

# 1 - d2h/dx2 
d2h_dx2 = sp.diff(dh_dx, x)

# 2 - d2h/dy2 
d2h_dy2 = sp.diff(dh_dy, y)

# 3 - Mixed second partial derivative d2h/dxdy
mix_dxy = sp.diff(dh_dx, y)

"""
Discussing symmetry: By Clairaut's Theorem, if If a function f(x,y) 
has continuous second partial derivatives in a region, then the mixed partial derivatives are equal. 
Therefore, in this case the mixed second partial derivative xy and yx would be equal
"""

#Show the results
print("\nSecond partial derivatives:")
print("∂²h/∂x²:", d2h_dx2)
print("∂²h/∂y²:", d2h_dy2)
print("∂²h/∂x∂y:", d2h_dx2)
```

```python
# (d) create a contour plot of the function j(x,y) = x3−3xy+y3

import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

#Define the function:
j = x**3- 3*x*y + y**3

# Convert the SymPy expression to a numerical function:
j_func = sp.lambdify((x, y), j, 'numpy')

#Create a grid of x and y values and evaluate j on this grid:
x_vals = np.linspace(-3, 3, 400)
y_vals = np.linspace(-3, 3, 400)
X, Y = np.meshgrid(x_vals, y_vals)
Z = j_func(X, Y)

#Plot the contour:
plt.contourf(X, Y, Z, levels=50, cmap='viridis')
plt.colorbar()
plt.title('Contour plot of $j(x, y) = x^3- 3xy + y^3$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.show()
```
